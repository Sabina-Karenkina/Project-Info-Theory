```python
def probability(my_dict):
    sum_symb = sum(my_dict.values())
    for i in my_dict:
        my_dict[i] = round(float(my_dict[i]/sum_symb),3)
    return my_dict
probability(symb_freq)
for j in symb_freq:
        print(j,'-', symb_freq[j])
```
